# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_path = Path("Resources/schools_complete.csv")
student_data_path = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_path)
student_data = pd.read_csv(student_data_path)

# Combine the data into a single dataset.  
school_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_df.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_df["school_name"].nunique()
school_count
```

```python
# Calculate the total number of students
student_totals = school_df["size"].unique()
student_count = student_totals.sum()
student_count
```

```python
# Calculate the total budget
school_budgets = school_df["budget"].unique()
total_budget = school_budgets.sum()
total_budget
```

```python
# Calculate the average (mean) math score
math_scores = school_df["math_score"].unique()
average_math_score = math_scores.sum() / len(math_scores)
average_math_score
```

```python
# Calculate the average (mean) reading score
reading_scores = school_df["reading_score"].unique()
average_reading_score = reading_scores.sum() / len(reading_scores)
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_df[(school_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
round(passing_math_percentage, 2)
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_df[(school_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
round(passing_reading_percentage, 2)
```

```python
# Use the following to calculate the percentage of students that passed math and reading
MR_passing_count = school_df[(school_df["math_score"] >= 70) & (school_df["reading_score"] >= 70)].count()["student_name"]
overall_passing_rate = MR_passing_count /  float(student_count) * 100
round(overall_passing_rate, 2)
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "Percent Passing Math": [passing_math_percentage],
    "Percent Passing Reading": [passing_reading_percentage],
    "Percent Overall Passing": [overall_passing_rate]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Percent Passing Math"] = district_summary["Percent Passing Math"].map("{:,.2f}%".format)


# Display the DataFrame
# district_summary = district_summary["Total Schools", "Total Students", "Total Budget", "Average Math Score", 
#                                     "% Passing Math", "% Passing Reading", "% Overall Passing"]
district_summary
```

## School Summary

```python
# Use the code provided to select the school type
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count
per_school_counts = school_data.set_index(["school_name"])["size"]
per_school_counts
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = school_df.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
print(per_school_capita)
print(per_school_budget)
```

```python
# Calculate the average test scores
per_school_math = school_df.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_df.groupby(["school_name"]).mean()["reading_score"]
passing_math_and_reading = school_df.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Calculate the number of schools with math scores of 70 or higher
school_passing_math = school_df[school_df["math_score"] >= 70].groupby(["school_name"])["math_score"].count()
school_passing_math
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = school_df[school_df["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
school_passing_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_passing_math / per_school_counts * 100
per_school_passing_reading = school_passing_reading / per_school_counts * 100
overall_passing_rate = (school_passing_math & school_passing_reading) / per_school_counts * 100
per_school_passing_math = per_school_passing_math.round()
per_school_passing_reading = per_school_passing_reading.round()
overall_passing_rate = overall_passing_rate.round()

print(per_school_passing_math.head())
print(per_school_passing_reading.head())
print(overall_passing_rate.head())
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "School Budget": per_school_budget,
    "Capita": per_school_capita,
    "Average Math Score": per_school_math.round(),
    "Average Reading Score": per_school_reading.round(),
    'Percent Passing Math': per_school_passing_math.map("{:,.2f}%".format),
    'Percent Passing Reading': per_school_passing_reading.map("{:,.2f}%".format),
    "Overall Passing Rate": overall_passing_rate.map("{:,.2f}%".format)})


# Formatting
per_school_summary["School Budget"] = per_school_summary["School Budget"].map("${:,.2f}".format)
per_school_summary["Capita"] = per_school_summary["Capita"].map("${:,.2f}".format)


# Display the DataFrame
per_school_summary
```
